###  Default Arguments

```python
def get_info(eid,ename,eage,org = 'BTS'):
    print('*******************************')
    print('Employee ID    :  ',eid)
    print('Employee Name  :  ',ename)
    print('Employee Age   :  ',eage)
    print('Company Name   :  ',org)
    print('*******************************')
```

```python
get_info(1,'Namjoon',29)
```

```python
def get_info2(eid,eage,org='BTS',*ename):                ## Variable length argument
    print('*******************************')
    print('Employee ID    :  ',eid)
    print('Employee Name  :  ',ename)
    print('Employee Age   :  ',eage)
    print('Company Name   :  ',org)
    print('*******************************')
```

```python
get_info2(1,29,'BTS','Namjoon','Jin','Suga','J-Hope','V','Jimin','JK')
```

###  Dictionary Format

```python
def get_info3(eid,eage,org='BTS',**ename):         ## keyword arguments       
    print('*******************************')
    print('Employee ID    :  ',eid)
    print('Employee Name  :  ',ename)
    print('Employee Age   :  ',eage)
    print('Company Name   :  ',org)
    print('*******************************')
```

```python
get_info3(1,29,'BTS',leader='Namjoon',vocal_1='Jin',rapper='Suga',dancer='J-Hope',vocal_2='V',dancer_2='Jimin',center='JK')
```

###  Variable length and keyword argument together

```python
def get_info4(eid,eage,*org,**ename):         ## keyword arguments       
    print('*******************************')
    print('Employee ID    :  ',eid)
    print('Employee Name  :  ',ename)
    print('Employee Age   :  ',eage)
    print('Company Name   :  ',org)
    print('*******************************')
```

```python
get_info4(1,29,'BTS','Big Hit','Hybe',leader='Namjoon',vocal_1='Jin',rapper='Suga',dancer='J-Hope',vocal_2='V',dancer_2='Jimin',center='JK')
```
